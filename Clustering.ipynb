```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv(r"data.csv", sep = ",")
data
```

```python
names = data["Name"]
names
```

```python
data["Name"] = pd.Categorical(data["Name"])
data["Name"] = data["Name"].cat.codes
data
```

```python
data_np = data.as_matrix()
data_np
```

# 3 clasters

```python
model = KMeans(n_clusters=3).fit(data_np)
```

```python
labels = model.labels_
centroids = model.cluster_centers_
```

```python
labels_unique = np.unique(labels)
n_clusters_ = len(labels_unique)

for k in range(n_clusters_):
    my_members = labels == k
    print ("cluster {0}: {1}".format(k, data_np[my_members, 0]))
    
```
